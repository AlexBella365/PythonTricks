# XKCD is fun!

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(150)/20
y = np.sin(x*np.pi)*x
```

```python
def printxy(x,y):
    plt.figure(figsize=(8,3))
    plt.plot(x,y)
    plt.plot(x,np.roll(y,10))
    plt.legend(['Normal','Shifted'])
```

## Normal

```python
plt.rcdefaults()
printxy(x,y);
```

## XKCD style

```python
plt.xkcd();
printxy(x,y);
```
